# Introduction

## Definition

There are multiple definitions of machine learning in different publications. But the general idea behind machine learning is that it is a technique in which the computer learns from the data and the performs a task without explicitly being programmed.

For formally the definition that describes machine learning best is this.

> A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E.
>
> *Tom Mitchell, 1997*

The most basic example of machine learning is the of detecting whether an email is spam or not (ham). The system uses existing data from the already existing emails which the user has manually labeled as spam and uses it to train itself. Now when a new email arrives it checks whether the email is similarly to that of already detected spam mail using various feature like likely words used in title, so same email id etc. Then decides (predicts) whether the new email is spam or not.

## Uses of Machine learning

Why use machine learning in the first place? Under what situations does machine learning helps?

### Preventing Complex Rule.

Let's take the email spam detection example. Suppose we have to build an spam filter using traditional programing approach. Then the ideal candidate would be to write a series of conditions using `if  else` branch. For example, if title of email contains words like "free", "amazing", "4U" or else if email is from a list of already identified email address etc. But as the spammer keeps updating his spamming technique then we will have to add new rule to the list. An soon the rule become unmanageable and complex.

Using machine learning we can let the system learn the rules on itself by comparing the data of existing classified email and keep learning on new email which were wrongly classified as safe but were actually spam.

### No traditional Solution.

Some computing problem are too difficult to solved by traditional procedure/rules approach. Consider the problem of speech recognition. To develop speech recognition software we would have to check each audio with combination of multiple world from a certain language and also multiple accents. Which is not possible to code.

The best solution is to write a algorithm which learns on itself.

### Discovering insight from data.

Machine learning can also help human learn insight from the data. Onces a machine learning algorithm has learned rule/patterns on itself then a human can inspect what the system has learned and get more insight from the data.

Applying Machine learning to discover patterns from huge data is called *data mining*.

## Types of Machine learning

Machine learning algorithms are broadly classified into three different categories.

- Whether or not they are trained with human supervision (supervised, unsupervised, semisupervised, and Reinforcement Learning)
- Whether or not they can learn incrementally on the fly (online versus batch learning)
- Whether they work by simply comparing new data points to known data points, or instead detect patterns in the training data and build a predictive model, much like scientists do (instance-based versus model-based learning)

### Supervised and Unsupervised Learning

Machine learning algorithms can be classified on the type of supervision required during the training phase.

#### Supervised Learning

In Supervised Learning, the data is supplied with label for training the algorithm.

Example of Supervised Learning is classification (Categorizing data) and regression (Predicting value from a given feature).

#### Unsupervised Learning

In Unsupervised learning the training data is unlabeled. The algorithm discover patterns or learns from data on it self.

Examples includes, Clustering algorithm to find unknown groups in data, Visualization algorithm to give a 2D or 3D visualization of data or Association Learning who's goal is to find unknown rules/association form a given data.

#### Semisupervised Learning

Some algorithm are combination of Supervised learning and Unsupervised learning.

The best example of this is Google Photos. The algorithm behind photos takes multiple photos and recognizes same person in multiple photos (Unsupervised) and then asks you to name that person and the assigned name to each present and future photo of that person with that name (Supervised).

#### Reinforcement Learning

In Reinforcement learning there is agent which performs some action in the environment and then receives feedback. The feedback is either negative or positive. And based on that feedback the agent either continues doing the think it is doing or improves its action.

Example is a game placing agent. For a chess playing agent the algorithm will play multiple games with itself and learn from the moves which result in higher score. Thus improving the moves as training improves over time.

### Online Vs Batch Learning

Other category is how the system learning. Some algorithms keeps learning continuously as the data is feed into the system. Example would be stock market prediction algo. The data stream is continuous for the algorithm and the learning of the system never stops. Such system is called Online system.

For some system the learning is incremental. Example can be Google speech recognition algorithm. Onces the model is trained the model is updated on the Android phones. And the team at Google would record more audios from different language and accents and the train the model and then later update the model again from the Android devices.

### Instance Vs Model Based Learning

Other way is how the algorithm learns from the data. One method is Instance based. In this method the algorithm has access to all the data and whenever a new instance is to categorized or predicted it check the other instances in the data which are similarly to that of new instance and assign the new instance to the same group of data or predict the value of new instance based on value of similar instance. Example would be spam detection, if a new email is 95% similar to existing spam email then possibility the email is spam.

Another way is that system build a model (equation) from the data and then use that model to predict the value for new instance. Example would be linear regression which tries to fit a line through the data and whenever a new instance is to be predicted it check where on the line the new instance fits and then predicts the value accordingly.

## Challenges in Machine Learning

There are some challenges with applying machine learning to get the right desire outcome from the data. Some of them are given below.

### Insufficient Quantity of Training Data.

When there is less data to train data the model would not fit to the ideal production data and thus the outcome would not be that accurate. It is sample as taking a small sample space and predicting the parameters of the population.

### Nonrepresentative Training Data.

Training data much be a good representative of future unknown instance. But the algorithm see a new data which is nothing similar to what it has seen in the training phase then the outcome is uncertain.

### Poor quality data.

Missing or bad data in training can also affect the outcome of the algorithm on live data.

### Irrelevant Features.

Not choosing a proper or relevant feature to predict the outcome will surely degrade the performance of the system. For example, will detecting spam email taking consideration whether it is raining will give curtain result.

### Overfitting Training Data.

Ever data has some noise in it, even training data. So creating a model which predicts the exact value for training data set is not a good idea that it has selected parameters which are too closely related to training data. Such model will performance badly on production data as the noise level will vary in production data.

### Underfitting Training Data.

Underfitting occurs when the model is too simple to learn the underlying structure of the data. For example using linear regression on a quadratic data.

## Testing and Validation

Onces training of the machine learning model is done how can we be sure that the model/algorithm will work on new unseen data inputs. The only way to make sure of this is to split the data into two sets. One set for training the model and the other to test the model. Usually the percentage is 80% as training data and 20% test data.

But there is a problem with this method. The problem is we use training data to train our system and then if the performance on test data fail we might over fit the model to the training data set. So another common practice is to divide the data into three sets training, validation and test. 

We train the model on training set so the model fits to the training dataset. Then we give it an unseen data set (validation set) and allow the model to change its parameter to based on validation set. And then we finally calculate the actual performance on the final test set.

Another practice to avoid wasting too much training data in validation set is to use cross-validation: the training set is split into complementary subsets, and each model is trained against a different combination of these subsets and validated against the remaining parts.

# Regression

Regression is a supervised, model based, batch algorithm. The algorithm creates a model/hypothesis from the data and then predict the value of new instance based on the learned hypothesis.

The flow of the regression algorithm is shown in the diagram below.

![Regression Flow](/static/img/notes/machine-learning/intro-to-machine-learning/regression_algo_flow.png)

The learning algorithm is given training data $X$ where $X$ can be a single input data or multiple input data. Input data is also called as *feature* and the output that we are trying to predict is called the *target*. A transitional example, would be to predict the prices of a house in a given area. In such case the prices of house might depend upon multiple parameters like size of the house, number of bedrooms, year constructed etc thus these become the *feature* of our training data set and the value we are trying to predict, price of the house, becomes the *target*.

The hypothesis $H(X)$ is a function of our feature which return the predict value of the target ($Y$). Given a training data to the learning algorithm it tries to build a hypothesis based on our initial assumption of the data. If the data is linear we will try to fit the a straight line through the data and then use the straight line to predict the price based on where the new feature meets the line. In such case the hypothesis will be a line function $Y=\theta_0 + X\theta_1$ where $\theta_0$ and $\theta_1$ are called the *parameters*. And the goal of the learning algorithm is to find the best values for the parameters will reduce our error in predicting the target value of a new unseen feature.

## Linear Regression

One example of regression algorithm is "Linear Regression". Lets take an example.

[stockprices.csv](https://raw.githubusercontent.com/sunitdeshpande/notes/master/static/raw/notes/machine-learning/intro-to-machine-learning/stockprices.csv)

Above is a stock prices of Google and Amazon from 04/02/2015 to 03/02/2016. We think that the there is an correlation between the stock prices of both of the stock prices. So we want to predict the price of stock value of Google (target) given stock value of Amazon (feature).

First lets see how this data looks.

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

CSV_FILE_PATH = '../../static/raw/notes/machine-learning/intro-to-machine-learning/stockprices.csv'

# Read CSV file
training_data = pd.read_csv(CSV_FILE_PATH, delimiter='\t')

# Plot data
plt.scatter(training_data.amazon, training_data.google)
plt.xlabel('Amazon')
plt.ylabel('Google')
plt.show()
```

As we can see from the data there is strong correlation between the data. I data almost follows a linear pattern. If we could fit a line through the data we can predict the price of stock price of Google given stock price of Amazon.

So the equation of line is given by $y=mx+c$ were $m$ is the slope of the line and $c$ is the y intercept of the line. Using this equation we can draw any line a 2D plane. And our data is also two dimensional (Amazon and Google price). To be consistent with machine learning algorithm we will write the equation as follows, $y=\theta_0 + x\theta_1$ where $\theta_0$ and $\theta_1$ are our parameters. So the goal of the linear regression algorithm is to find the best possible value for $\theta_0$ and $\theta_1$ such that we get a line which perfectly fits the data. Thus our hypothesis in this example can be written as.

$$
h(x) = h_\theta(x) = \theta_0 + x \theta_1
$$

The above definition get a little big and complex when we start working with multiple parameters and features. So we can reduce the definition using some vector notions.

We can denote the parameters of the algorithm as vector of $M$ rows (also called as parameter vector) where $M$ denotes the number of training data (number of rows in csv for this example). And let $N$ denote the total number of features. Then

$$
\theta = \begin{bmatrix} \theta_0 \\ \theta_1 \\ - \\ - \\ \theta_N \end{bmatrix}
$$

And if we can denote features such $x_1, x_2, x_3, ... , x_N$ and let $x_0=1$ then we can write the feature as a vector of $N$ rows (also called as feature vector) as follows.

$$
X = \begin{bmatrix} x_0 \\ x_1 \\ - \\ - \\ x_N \end{bmatrix}
$$

Thus our Hypothesis function becomes.

$$
h(x) = \begin{bmatrix} \theta_0, \;  \theta_1, \;  -, \;  -, \;  \theta_N \end{bmatrix} * \begin{bmatrix} x_0 \\ x_1 \\ - \\ - \\ x_N \end{bmatrix} = \theta^T X
$$

### Cost function

How do we decide what values we use for our parameters ($\theta$) so that our line fits perfectly to the data?

If we think what we are trying to do here is we already have some know values of our target $Y$ (Google stock price) for some know feature value $X$ (Amazon stock price). And we want to select some value for our parameters ($\theta$) such that the value of our hypothesis function ($h(x)$) is as close to the value of our target $Y$. So how we decide which value of parameter is good or bad.

One way to do this is by calculating the predicted value $Y$ for know $X$ and then taking difference between them to see how are they from each other. Then the mean of the differences will give us an average of how far are the predicted values from the actual values.

$$
\frac{1}{M}\sum_{i=0}^M (h(x^{(i)}) - y^{i})
$$

Where $x^{(i)}$ denotes the $i^{th}$ feature and $y^{(i)}$ denotes the $i^{th}$ target value.

But the problem here is that some difference would be positive and others would be negative and they might cancel out each other. To overcome this we take the square of the difference of individual instance and then add them, thus giving us.

$$
\frac{1}{M} \sum_{i=0}^M (h_\theta(x^{(i)}) - y^{i})^2
$$

This above way to calculating the error in the prediction is called the **Mean Square Error**.

We can generalize the above equation as a function of $\theta$ as follows.

$$
J(\theta) = \frac{1}{M} \sum_{i=0}^M (h_\theta(x^{(i)}) - y^{i})^2
$$

The above function $J(\theta)$ is generally called the "cost function". And the goal of the learning algorithm is to choose value of our parameters ($\theta$) and minimize $J(\theta)$ such that $J$ gets as close as possible to zero ($J \to 0$).

But for linear generation we multiple the function by $\frac{1}{2}$ so to simply further calculation (in gradient descent). So for linear regression the cost function is defined as below.

$$
J(\theta) = \frac{1}{2M} \sum_{i=0}^M (h_\theta(x^{(i)}) - y^{i})^2
$$


```python
%matplotlib inline

import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

CSV_FILE_PATH = '../../static/raw/notes/machine-learning/intro-to-machine-learning/stockprices.csv'

# Cost function
def get_cost(amazon, google, theta_0, theta_1):
    # Hypothesis function
    predicted_values = amazon * theta_1 + theta_0

    # Difference between predicted value and target
    difference = google - predicted_values
    
    # Square of difference
    sum_of_square_difference = difference ** 2

    # Sum of square of difference
    sum_of_difference = np.sum(sum_of_square_difference)
    
    # sum_of_difference multipled by 1/2m.
    return sum_of_difference * (1/(2*len(amazon)))


# Read CSV file
training_data = pd.read_csv(CSV_FILE_PATH, delimiter='\t')

# Read CSV file
training_data = pd.read_csv(CSV_FILE_PATH, delimiter='\t')

# Normalize Data
amazon = (training_data.amazon-400)/(1000-400)
google = (training_data.google-400)/(1000-400)

# Theta values
theta_0_values = np.arange(-2.0, 2.0, 0.1)
theta_1_values = np.arange(-2.0, 2.0, 0.1)

# Compute Cost functon
xx, yy = np.meshgrid(theta_0_values, theta_1_values)
costs = np.array([get_cost(amazon, google, a, b) for a, b in zip(np.ravel(xx), np.ravel(yy))])

# Plot cost function
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

Z = costs.reshape(xx.shape)
ax.plot_surface(xx, yy, Z)

ax.set_xlabel('Theta 0')
ax.set_ylabel('Theta 1')
ax.set_zlabel('Cost Function')

plt.show()
```

In the above diagram we have plotted the cost function of our data with linear hypothesis. As we can see the cost the function is near zero when $\theta_0$ is between $0.5$ and $1.0$ and $\theta_1$ between $-1$ and $-0.5$.

## Gradient Descent

We have seen how the we can compute the ideal value of the parameters from plotting data. But how do we compute the value of the parameters algorithmically?

There are multiple ways of doing so but the most common way of doing it is "Gradient Descent".

The idea of gradient descent is simple, we select some random point on our cost function plot and then check our surrounding in 360 degree to see going in which direction will help us reduce the value of function.

![Gradient Descent](static/img/notes/machine-learning/intro-to-machine-learning/gradent_decent.png)

In the above diagram we can see that we start from a random point uphill and then we check in which direction should we go down hill and then move it in that direction and continue the steps until we reach a local minimum where there is no place to go. And that becomes our value of parameters.

So lets say we are at some random point on the function, how do we decide in which direction should be go to minimize the function. One way to do this is to take derivative of the function.

![Gradient Descent Derivative](static/img/notes/machine-learning/intro-to-machine-learning/gradient_descent_derivative.png)

As shown in above diagram, if we take derivate of a function at a particular point it would give us the slope of the line which is tangent to the function at that point. Because derivate is given as rate of change which is called as $\frac{f(x+h) - f(x)}{\Delta h}$ where $\Delta h$ represents small change in $x$. When $\Delta h$ is very small the value of derivate is same as the slope of the tangent of the point $\frac{\Delta y}{\Delta x}$. Also the slope of the line gives us the direction of the steepness. As seen from the diagram if the slope is positive then we must decrease the value of parameter and if it is negative then we need to increase the value of our function to minimize the cost function.

Thus to calculate our next value of $\theta$ we can use the following formulae.

$$
\theta := \theta - \frac{\partial}{\partial\theta} J(\theta)
$$

Derivation of partial derivative of the cost function goes as follows.

Lets find derivate for $\theta = \theta_0$

$$
\begin{align*}
\frac{\partial}{\partial\theta_0} J(\theta) &= \frac{\partial}{\partial\theta_0} \frac{1}{2m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)})^2 \\
&= \frac{1}{2m} \frac{\partial}{\partial\theta_0} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)})^2 \tag{Scalar multiplication} \\
&= \frac{1}{2m} \sum_{i=0}^{m} \frac{\partial}{\partial\theta_0} (h_{\theta}^{(i)} - y^{(i)})^2  \tag{Sum Rule} \\
&= \frac{1}{2m} \sum_{i=0}^{m} 2 (h_{\theta}^{(i)} - y^{(i)})  \frac{\partial}{\partial\theta_0} (h_{\theta}^{(i)} - y^{(i)}) \tag{Chain Rule} \\
&= \frac{2}{2m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)})  \frac{\partial}{\partial\theta_0} (\theta_0 + x^{(i)}\theta_1 + ... + x^{(n)} \theta_n - y^{(i)}) \tag{By definition}  \\
\frac{\partial}{\partial\theta_0} J(\theta) &= \frac{1}{m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)}) \\
\end{align*}
$$

In the last step as we are taking derivate with respect to $\theta_0$ everything is constant expect $\theta_0$ and $\frac{\partial}{\partial\theta_0} \theta_0 = 1$. Thus giving the value of derivate as $1$.

Now for some other value of $\theta_j$ expect $\theta_0$.

$$
\begin{align*}
\frac{\partial}{\partial\theta_j} J(\theta) &= \frac{\partial}{\partial\theta_j} \frac{1}{2m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)})^2 \\
&= \frac{1}{2m} \frac{\partial}{\partial\theta_j} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)})^2  \\
&= \frac{1}{2m} \sum_{i=0}^{m} \frac{\partial}{\partial\theta_j} (h_{\theta}^{(i)} - y^{(i)})^2  \\
&= \frac{1}{2m} \sum_{i=0}^{m} 2 (h_{\theta}^{(i)} - y^{(i)})  \frac{\partial}{\partial\theta_j} (h_{\theta}^{(i)} - y^{(i)})  \\
&= \frac{2}{2m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)})  \frac{\partial}{\partial\theta_j} (\theta_0 + x^{(1)}\theta_1 + ... + x^{(n)} \theta_n - y^{(i)})  \\
\frac{\partial}{\partial\theta_j} J(\theta) &= \frac{1}{m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)}) x^{(j)} \\
\end{align*}
$$

In the last step as we are taking derivate with respect to $\theta_i$ everything is constant expect $\theta_i$. Thus the equation becomes $\frac{\partial}{\partial\theta_i} \theta_ix^{(i)} = x^{(i)}$.


Thus our final equation for calculation the next value of $\theta$ becomes.

$$
\begin{aligned}
\theta_0 &:= \theta_0 - \frac{1}{m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)}) \\
\theta_j &:= \theta_j - \frac{1}{m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)}) x^{(j)}
\end{aligned}
$$

To give a manual control over the step size given by the slope of the line (derivative). We generally multiply the derivate by a number called as learning rate and denoted by $\alpha$. By increasing the $\alpha$ we can increase the number of step the next $\theta$ jumps to and opposite if we reduce alpha. So the final equation that is generally used in practice is as follows.

$$
\begin{aligned}
\theta_0 &:= \theta_0 - \frac{\alpha}{m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)}) \\
\theta_j &:= \theta_j - \frac{\alpha}{m} \sum_{i=0}^{m} (h_{\theta}^{(i)} - y^{(i)}) x^{(j)}
\end{aligned}
$$

Choosing the learning rate must be done properly, because if the learning rate is chosen very high the the algorithm will take forever to reach the local minimum. And if the learning rate is chosen very high the algorithm will keep jumping from positive to negative slope and never reach the local minimum. Best idea to check cost function by starting with learning rate as $0.5, 0.1, 0.01, 0.001$ etc.

### Types of gradient descent.

When using gradient descent in practical scenario there are computational challenges. Which include the time it take to calculate the next value of $\theta$ or the amount of data which can used for computing at a single instances of time. So to overcome such challenges there variation of the algorithm have been developed.

- Batch Gradient descent.
- Stochastic Gradient Descent.
- Mini Batch Gradient Descent
- Coordinate Gradient Descent.

#### Batch Gradient descent.

The algorithm we saw above is the batch gradient descent. In which to find the next value of parameters the entire training set is needed. Then for next iteration again the parameters are updated and entire training data is used to calculate the next value of the parameter. Each such iteration is generally called to *epoch*. This will result in accurate values for our parameter but is very slow as we iterate over our entire training set.

#### Stochastic Gradient Descent.

To work on drawback of batch gradient descent. Stochastic take one instance of training data at a time and adjusts the parameter based on the outcome of a single training instance. It will not result in an accurate parameter value but it will be quicker as only one iteration over training set is required.

#### Mini Batch Gradient Descent

Mini Batch algorithm is a combination of batch and stochastic gradient descent. Initially the training data is divided into multiple smaller sets generally between 50 to 200 instances. And for each iteration we choose and random set from the sets of training data we partition earlier. Mini batch algorithm is more faster but less accurate than batch gradient decent and less faster but more accurate than stochastic gradient descent.

#### Coordinate Gradient Descent.

Another variation of gradient descent is Coordinate. In this algorithm instead of computing gradient decent on each feature we choose only one feature at a time for a single iteration. The and move in the direction of slope for that feature. The choose of feature can be round robin or random. But the idea behind it is that if we choose single feature at a time we will eventually reach the global minium of all the feature.

## Closed - Form solution

Another way to minimize the cost function is to take the derivative of the cost function and set it to zero. Because the function will have a zero derivative at local minimum. And then we can use that derivative to compute the value of parameters which minimizes the cost function.

To calculate the derivative of cost function we will introduce some notions.

Let $X$ denote matrix ($X \rightarrow \mathbb{R}^{N*M}$). Where $N$ is the number of feature and $M$ be the number of training samples. $X$ contains all training samples.

Let $\theta$ be a vector of parameters and $y$ be a vector of target values.

Then we can write cost function as follows.

$$
J(\theta) = \frac{1}{2m} (X\theta - y)^T (X\theta - y)
$$

We take square of difference as before by using inner product of the difference.

We can simplify the equation as follows.

$$
\begin{aligned}
J(\theta) &= \frac{1}{2m} (X\theta - y)^T (X\theta - y) \\
&= \frac{1}{2m} ( \; ( \; (X\theta)^T - y^T \; ) (X\theta - y) \; ) \\
J(\theta) &= \frac{1}{2m} (\; (X\theta)^TX\theta - (X\theta)^T y - y^TX\theta -y^Ty  \;) \\
\end{aligned}
$$

Now here $X\theta$ and $y$ are vectors so it doesn't matter in which order we multiple them as long as the dimension matches. So $(X\theta)^Ty = y^TX\theta$. Thus our equation becomes,

$$
J(\theta) = \frac{1}{2m} (\; (X\theta)^TX\theta - 2(X\theta)^T y -y^Ty  \;)
$$

Now we take partial derivative of the function with respect to $\theta$. Derivate of a matrix/vector is similar to that of scalar quantity.That is when we take derivative of a vector/matrix we take derivative of each individual element of the matrix.

$$
\begin{aligned}
\frac{\partial J}{\partial\theta}  &= \frac{\partial}{\partial\theta} \frac{1}{2m} (\; (X\theta)^TX\theta - 2(X\theta)^T y -y^Ty  \;) \\
&= \frac{1}{2m} \frac{\partial}{\partial\theta} (\; (X\theta)^TX\theta - 2(X\theta)^T y -y^Ty  \;) \\
&= \frac{1}{2m} (\;
\frac{\partial}{\partial\theta}( (X\theta)^TX\theta ) -  \frac{\partial}{\partial\theta} 2(X\theta)^T y - \frac{\partial}{\partial\theta} y^Ty  \;) \\
&= \frac{1}{2m} (\;
\frac{\partial}{\partial\theta}( (X\theta)^TX\theta ) -  \frac{\partial}{\partial\theta} 2(X\theta)^T y  \;) \\
\frac{\partial J}{\partial\theta}  &= \frac{1}{2m} (\;
\frac{\partial}{\partial\theta}( X^T\theta^TX\theta ) -  2X^T y  \;) \\
\end{aligned}
$$

We if we consider the first equation.

$$
\begin{aligned}
\frac{\partial}{\partial\theta}( X^T\theta^TX\theta ) &= X^T X \frac{\partial}{\partial\theta} (\theta^T \theta)
\end{aligned}
$$

We have a rule of derivate of vectors.

$$
\frac{\partial x^Tx}{\partial x} = 2 x
$$

We can understand the intuition behind the rule as follows,

$$
\begin{aligned}
x &= \begin{bmatrix} x_1 \\ x_2 \\ - \\ - \\ x_n \end{bmatrix}\\
x x^T &= \begin{bmatrix} (x_1)^2 \\ (x_2)^2 \\ - \\ - \\ (x_n)^2 \end{bmatrix} \\
\frac{\partial x x^T}{\partial x} &= \begin{bmatrix} 2x_1 \\ 2x_2 \\ - \\ - \\ 2x_n \end{bmatrix} \\
\frac{\partial x x^T}{\partial x} &= 2\begin{bmatrix} x_1 \\ x_2 \\ - \\ - \\ x_n \end{bmatrix} \\
\end{aligned}
$$

Thus our derivation becomes.

$$
\begin{aligned}
\frac{\partial}{\partial\theta}( X^T\theta^TX\theta ) &= X^T X \frac{\partial}{\partial\theta} (\theta^T \theta) \\
&= X^T X 2 \theta
\end{aligned}
$$

Substituting the above value in derivative of $J(\theta)$ we get.

$$
\begin{aligned}
\frac{\partial J}{\partial\theta} &= \frac{1}{2m} (\;
X^T X 2 \theta -  2X^T y  \;) \\
&= \frac{1}{2m} (\; 2X^T X \theta -  2X^T y  \;)
\end{aligned}
$$

Now to find the local minimum of the function we set derivate to zero.

$$
\begin{aligned}
\frac{\partial J}{\partial\theta} = \frac{1}{2m} (\; 2X^T X \theta -  2X^T y  \;) &= 0 \\
2X^T X \theta -  2X^T y &= 0 \\
2X^T X \theta &=  2X^T y \\
X^T X \theta &=  X^T y \\
\end{aligned}
$$

Now assuming that $(X^TX)$ is invertible. We can write.

$$
\theta = (X^TX)^{-1}X^Ty
$$

The above equation is called "normal equation" or the "closed form equation". Which can be used to directly compute the value of the parameters.

Closed form linear regression is only useful if number of samples is less than $1000$ because it involves matrix multiplication which is an $O(n^3)$ costly operation. For bigger data the normal iterative gradient descent is preferred.

## Computing Linear Regression

### Scikit Learn

```python
import pandas as pd
import numpy as np
from sklearn import linear_model

CSV_FILE_PATH = '../../static/raw/notes/machine-learning/intro-to-machine-learning/stockprices.csv'

values_to_predict = np.array([400, 600, 512.35, 650.12]).reshape(-1, 1)

# Read CSV
training_data = pd.read_csv(CSV_FILE_PATH, delimiter='\t')

# Arrange data as required by scikit learn
X = training_data.amazon.to_numpy().reshape(-1, 1)
Y = training_data.google.to_numpy().reshape(-1, 1)

# Train model (No test data :P)
model = linear_model.LinearRegression()
model.fit(X, Y)
print('Model Information...')
print('Theta 0 / Slope = {}'.format(model.coef_[0][0]))
print('Theta 1/ Intercept = {}'.format(model.intercept_[0]))
print()

# Predict Value
predicted_values = model.predict(values_to_predict)

print('Predictions....')
for value, prediction in zip(values_to_predict, predicted_values):
    print('{} -> {}'.format(value[0], prediction[0]))
```
